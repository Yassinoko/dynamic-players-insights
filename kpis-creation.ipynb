```python
goals_forward = round(df_goals[df_goals['player_name'] == player_name]['goals'].sum(),2)
goals_forward
```

```python
goals_forwardpergame = round(df_goals[df_goals['player_name'] == player_name]['goals'].sum() / df_attempts[df_attempts['player_name'] == player_name]['match_played'].sum(),2)
goals_forwardpergame
```

```python
assists_forward = round(df_attacking[df_attacking['player_name'] == player_name]['assists'].sum(),2)
assists_forward
```

```python
assists_forwardpergame = round(df_attacking[df_attacking['player_name'] == player_name]['assists'].sum() / df_attempts[df_attempts['player_name'] == player_name]['match_played'].sum(),2)
assists_forwardpergame
```

```python
shots_on_target_forward = round(df_attempts[df_attempts['player_name'] == player_name]['on_target'].sum(),2)
shots_on_target_forward
```

```python
shots_on_target_pergame_forward = round(df_attempts[df_attempts['player_name'] == player_name]['on_target'].sum() / df_attempts[df_attempts['player_name'] == player_name]['match_played'].sum(),2)
shots_on_target_pergame_forward
```

```python
successful_dribbles_forward = round(df_attacking[df_attacking['player_name'] == player_name]['dribbles'].sum(),2)
successful_dribbles_forward
```

```python
output_forward = {
    "Total goals" : goals_forward, 
    "Goals per game" : goals_forwardpergame,
    "Total assists" : assists_forward, 
    "Assists per game" : assists_forwardpergame,
    "Shots on target" : shots_on_target_forward,
    "Shots on target per game" : shots_on_target_pergame_forward,
    "Successful dribbles" : successful_dribbles_forward}
output_forward
```

```python
def calculate_forward_kpis(player_name):
    return output_forward
```
